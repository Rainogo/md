# PreparaciÃ³n de Datos

```python
datos.head()
```

```python
datos.info()
```

```python
pd.get_dummies(datos)
```

```python
X = pd.get_dummies(datos)
```

```python
X.describe()
```

```python
X.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
escala = StandardScaler()
```

```python
escala.fit(X)
```

```python
X_escala = escala.transform(X)
```

```python
X_escala
```

# CreaciÃ³n del Modelo KMeans

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=2)
```

```python
modelo.fit(X_escala)
```

```python
etiqueta_grupo = modelo.predict(X_escala)
```

```python
etiqueta_grupo
```

```python
len(etiqueta_grupo)
```

```python
X['Grupo'] = etiqueta_grupo
```

```python
X.head()
```

```python
plt.figure(dpi=200)
sns.heatmap(X.corr())
```

```python
X.corr()['Grupo']
```

```python
plt.figure(figsize=(12,6),dpi=200)
X.corr()['Grupo'].iloc[:-1].sort_values().plot(kind='bar')
```

# ElecciÃ³n del K adecuado

```python
ssd = []
for k in range(2,15):
    modelo = KMeans(n_clusters = k)
    modelo.fit(X_escala)
    ssd.append(modelo.inertia_)
```

```python
plt.plot(range(2,15),ssd,'o--')
plt.xlabel("Valor de K")
plt.ylabel("Suma de distancias al cuadrado (ssd)")
```

AnÃ¡lisis de la reducciÃ³n SSE

```python
ssd
```

```python
2469792.361662749-2370786.871245429
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```
